# Script launcher

```python
import subprocess
```

```python
# files
DIRPATH_MODIF = "../../"
DIRPATH_PERFORMANCE_TEST = "phdtrack_data/Performance_Test/Performance_Test/"
DIRPATH_VERSION = "V_7_1_P1/"
DIRPATH_BYTES = "16/"

FILES = ["10535-1650982250"]
```

## Decrypt a single pcap file

```python
PATH = DIRPATH_MODIF + DIRPATH_PERFORMANCE_TEST + DIRPATH_VERSION + DIRPATH_BYTES + FILES[0]
PATH_PCAP = PATH + ".pcap"
PATH_JSON = PATH + ".json"
args = ("go", "run", "./single-decrypt/decrypt.go", "ctr", PATH_PCAP, PATH_JSON)
popen = subprocess.Popen(args, cwd="../ssh_key_bruteforcer_and_decryptor/")
popen.wait()
```

## Bruteforce a single pcap file -- Key A & C

```python

PATH = DIRPATH_MODIF + DIRPATH_PERFORMANCE_TEST + DIRPATH_VERSION + DIRPATH_BYTES + FILES[0]
PATH_PCAP = PATH + ".pcap"
PATH_HEAP = PATH + "-heap.raw"
args = ("go", "run", "./naive-single-bruteforce/bruteforce.go", "ctr", PATH_PCAP, PATH_HEAP, str(DIRPATH_BYTES).replace("/", ""))
popen = subprocess.Popen(args, cwd="../ssh_key_bruteforcer_and_decryptor/")
popen.wait()
```
